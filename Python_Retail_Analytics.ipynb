###

### Creating Subset of Data

```python
#Creating subset for campaign acceptance
campaigns = retail[['ID', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']]
campaigns.head()
```

```python
#Substituting values in subset with 'Yes' for 1 and 'No' for 0
campaigns['AcceptedCmp1'] = campaigns['AcceptedCmp1'].replace(1, 'Yes')
campaigns['AcceptedCmp1'] = campaigns['AcceptedCmp1'].replace(0, 'No')
campaigns['AcceptedCmp2'] = campaigns['AcceptedCmp2'].replace(1, 'Yes')
campaigns['AcceptedCmp2'] = campaigns['AcceptedCmp2'].replace(0, 'No')
campaigns['AcceptedCmp3'] = campaigns['AcceptedCmp3'].replace(1, 'Yes')
campaigns['AcceptedCmp3'] = campaigns['AcceptedCmp3'].replace(0, 'No')
campaigns['AcceptedCmp4'] = campaigns['AcceptedCmp4'].replace(1, 'Yes')
campaigns['AcceptedCmp4'] = campaigns['AcceptedCmp4'].replace(0, 'No')
campaigns['AcceptedCmp5'] = campaigns['AcceptedCmp5'].replace(1, 'Yes')
campaigns['AcceptedCmp5'] = campaigns['AcceptedCmp5'].replace(0, 'No')
campaigns['Response'] = campaigns['Response'].replace(1, 'Yes')
campaigns['Response'] = campaigns['Response'].replace(0, 'No')

campaigns.head()
```
